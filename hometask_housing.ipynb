## ДЗ №1 по ML. Предсказать параметр MEDV из датасета housing.csv:
#### 1 - модель линейной регрессии 
#### 2 - разбить train/test на 0.8/0.2

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
RANDOM_SEED = 777
```

```python
# чтение файла
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# делим на train/test выборки
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis = 1), 
                                                    df['MEDV'], 
                                                    test_size = 0.2,
                                                   random_state = RANDOM_SEED)
```

```python
X_train
```

```python
X_test
```

```python
model = LinearRegression()
```

```python
# обучаем модель
model.fit(X_train,y_train)
```

```python
pred = model.predict(df.drop('MEDV', axis = 1))
```

```python
# добавляем поле с предсказанием
df['predict'] = pred
```
