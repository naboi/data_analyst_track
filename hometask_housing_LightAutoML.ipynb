```python
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
from sklearn.model_selection import train_test_split
```

```python
RANDOM_SEED = 777
```

```python
# чтение файла
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# делим на train/test выборки
train_data, test_data = train_test_split(df, 
                                       test_size=0.2,
                                       random_state=RANDOM_SEED)
```

```python
train_data
```

```python
y_train = train_data['MEDV']
y_test = test_data['MEDV']
```

```python
task = Task(name='reg' ) # регрессия
roles = {'target': 'MEDV'} # целевой признак - MEDV
automl = TabularAutoML(task=task)
oof_pred = automl.fit_predict(train_data, roles=roles)
```

```python
train_data = train_data.drop('MEDV',axis=1)
test_data = test_data.drop('MEDV',axis=1)
```

```python
y_train_score = automl.predict(train_data).data
y_test_score = automl.predict(test_data).data
```

```python
y_test_score = [val for sublist in y_test_score for val in sublist]

results = pd.DataFrame()
results['MEDV'] = y_test
results['score'] = y_test_score
```

```python
results
```

```python
# посчитаем точность
errors = mean_squared_error(results['MEDV'], results['score'], squared=False)
print(errors)
```
